## [0] 그래프그리기

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy as sp
np.random.seed(0)
N1 = 50
N2 = 100
x1 = sp.stats.norm(0, 1).rvs(N1)
x2 = sp.stats.norm(0.5, 1.5).rvs(N2)
plt.figure(figsize=(12,8))
ax = sns.distplot(x1, kde=False, fit=sp.stats.norm, label="data set 1")
ax = sns.distplot(x2, kde=False, fit=sp.stats.norm, label="data set 2")
ax.lines[0].set_linestyle(":")
plt.legend()
plt.show()
```

## [1] 정규성 검정

### (1)  Shapiro-Wilk test for normality
+ [공식문서](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.shapiro.html)

```python
from scipy import stats
import numpy as np
rng = np.random.default_rng()
x = stats.norm.rvs(loc=5, scale=3, size=100, random_state=rng)
shapiro_test = stats.shapiro(x)
shapiro_test
```

```python
shapiro_test.statistic
shapiro_test.pvalue
```

## [2] 등분산 검정

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy as sp
N1 = 100
N2 = 100
sigma_1 = 1
sigma_2 = 1.2
np.random.seed(0)
x1 = sp.stats.norm(0, sigma_1).rvs(N1)
x2 = sp.stats.norm(0, sigma_2).rvs(N2)

plt.figure(figsize=(12,8))
ax = sns.distplot(x1, kde=False, fit=sp.stats.norm, label="data set 1")
ax = sns.distplot(x2, kde=False, fit=sp.stats.norm, label="data set 2")
ax.lines[0].set_linestyle(":")
plt.legend()
plt.show()
```

```python
sp.stats.bartlett(x1, x2)
```

```python
sp.stats.fligner(x1, x2)
```

```python
sp.stats.levene(x1, x2)
```

## [3] 독립표본 t-test
+ [공식문서](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_ind.html)

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy as sp
N_1 = 10
mu_1 = 0
sigma_1 = 1
N_2 = 10
mu_2 = 0.5
sigma_2 = 1
np.random.seed(0)
x1 = sp.stats.norm(mu_1, sigma_1).rvs(N_1)
x2 = sp.stats.norm(mu_2, sigma_2).rvs(N_2)

plt.figure(figsize=(12,8))
ax = sns.distplot(x1, kde=False, fit=sp.stats.norm, label="data set 1")
ax = sns.distplot(x2, kde=False, fit=sp.stats.norm, label="data set 2")
ax.lines[0].set_linestyle(":")
plt.legend()
plt.show()
```

```python
sp.stats.ttest_ind(x1, x2, equal_var=False)
```
