+ `.astype(str)` 넣어줘야, AttributeError: Can only use .str accessor with string values 해결가능

```python
# 정규표현식 처리 가능! - 매우 빠름
df['title'] = df['title'].astype(str).str.replace("[^ㄱ-ㅎ ㅏ-ㅣ 가-힣 a-z A-Z]", " ")

df['new_topic'] = df['new_topic'].astype(str).str.replace("문화|생활", "문화/생활")  # 문화 or 생활을  -> 문화/생활 로 바꾸기
```
