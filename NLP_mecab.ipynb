```python
stop_words = set(['불용어1','불용어2'])
from konlpy.tag import Mecab
mecab = Mecab()

# 1.string에서 noun만 뽑기
def get_nouns(sentence):
    nouns = []
    nouns.append(' '.join([noun for noun in mecab.nouns(str(sentence))
                        if noun not in stop_words and len(noun) > 1])) # 불용어처리 & 1글자 제외
    return nouns

print(get_nouns('한글 문장의 명사를 뽑는 코드 인데 만족스러우신가요? '))
```
