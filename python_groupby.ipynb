```python
# groupby sample


# groupby sampling : 최소개수 선택가능. 
## groupby.sample과 차이는,  groupby.sample은 최소 그룹 개수를 최대값으로 샘플링 가능함
## 이거는 최소 개수가 설정 가능함
def sampling_func(data, sample_pct):
    np.random.seed(1)
    N = min(200,len(data))
    sample_n = int(N*sample_pct) # integer
    sample = data.take(np.random.permutation(N)[:sample_n])
    return sample

from tqdm.notebook import tqdm
tqdm.pandas()

sample_df = df.groupby('new_topic', group_keys=False).progress_apply(sampling_func, sample_pct=1)
```
