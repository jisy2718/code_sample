### (1) key, value 모두 이용한 정렬

+ dict = { '키1' : 5 , '키2' :10 , ... } 과 같은 꼴
    + value가 큰 것이 앞으로 오도록하고, value가 같다면, 글자 길이 짧은 것이 앞에오도록 정렬

```python
total_kwd_dict = {'로나':1232, '코로':1241, '코로나':1100, '제니스':4002, '제니':4213}
### (1) 글자가 짧은 것이 앞에 오도록 정렬
sorted_kwd_dict_lst = sorted(total_kwd_dict.items(), key=lambda item: len(item[0]), reverse=False)
### (2) count가 많이 된 것이 앞에 오도록 정렬
sorted_kwd_dict_lst = sorted(sorted_kwd_dict_lst, key=lambda item: item[1], reverse=True)
```
