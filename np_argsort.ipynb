### (1) argsort

+ argsort 이용해서, 2개의 list를 대응시켜서 sorting 가능

```python
import numpy as np

a = [1,4,2,3,4,5]
b = ['a','b','c','d','e']
# b.sorted(key=lambda x :)
print(np.argsort(a, ))
b_sorted = [b[idx] for idx in np.argsort(a)]
b_sorted
```

```python
a = [100,100,100,200,300,400]
b = ['a','b','c','d','e','f']

no_dup = [ idx for idx, value in enumerate(a) if idx == a.index(value)]
no_dup
```

```python
b_sorted = [b[idx] for idx in no_dup]
b_sorted
```
