### .values

```python
import pandas as pd

a = pd.DataFrame({'col1':[1,2,3],'col2':[11,111,33]})
print(a.values)

print('-----')

for col1, col2 in a.values:
    print(col1, col2)
```

### .loc
+ index으로 접근
+ pandas_df.loc[:n] : n까지 indexing

```python
import pandas as pd

a = pd.DataFrame({'col1':['첫째행','둘째행','셋째행'],'col2':[1,2,3]}, index=(0,1,3))
a
```

```python
a.loc[:1]
```

```python
a.loc[[1]] # 1개 row 접근할 때, df로 만들고 싶으면 [] 2개 이용
```

```python
a.loc[1]
```

```python
a.loc[2]
```

### .iloc
+ index가 아닌 순서로 접근
+ pandas_df.iloc[:n] -> n은 제외하고 indexing

```python
import pandas as pd

a = pd.DataFrame({'col1':['첫째행','둘째행','셋째행'],'col2':[1,2,3]}, index=(0,1,3))
a
```

```python
a.iloc[2] # index 접근이 아니라, (0부터 시작해서) 2번째 행에 접근하는 것
```

```python
a.iloc[:3]
```

```python
a.iloc[:2]
```

```python
a.iloc[:0]
```

```python
a.iloc[[1]]
```

```python
a.iloc[1]
```
