```python
# 결과 확인 : acc 높은 순서대로 parameter와 함께 정렬됨
rank_test_score = sgd_cv.cv_results_['rank_test_score'].tolist()  # 각 model들의 순위가 1부터 표기되어 있음 [3,1,2,4] 이런 식이면 두번재 모델이 1등이란 소리
mean_test_score = sgd_cv.cv_results_['mean_test_score'].tolist()   # [0.8, 0.9, 0.85, 0.75] 와 같음. 각 모델의 score가 나타나있음
mean_test_score = [round(score,4) for score in mean_test_score]

params = sgd_cv.cv_results_['params']                             # 각 model들의 파라미터 정보를 담고 있음

results = list(zip(rank_test_score,mean_test_score,params))
results.sort(key=lambda x : x[0])

for result in results:
    print(result)
```
