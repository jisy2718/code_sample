## [2] 인터넷에서 본 간단한 예시함수

### (1) 적용 x

```python
import time
def do_something():
    print('1초간 잠을 잡니다...')
    time.sleep(1)
    print('잠에서 깨었습니다...')
 
if __name__ == '__main__':
 
    start = time.perf_counter()
 
    for _ in range(2):
        do_something()
    
    finish = time.perf_counter()
 
    print(f'{round(finish-start,2)}초 만에 작업이 완료되었습니다.')
```

### (2) 적용 o

```python
import time
import multiprocessing
 
def do_something(arg1, arg2):
    print('1초간 잠을 잡니다...')
    print(arg1, arg2)
    time.sleep(1)
    print('잠에서 깨었습니다...')
 
if __name__ == '__main__':
 
    start = time.perf_counter()
 
    processes = []
    for _ in range(2): # multi process 개수 정하기
        p = multiprocessing.Process(target=do_something, args=(10,11)) ## 각 프로세스에 작업을 등록
        p.start()
        processes.append(p)
 
    for process in processes:
        process.join()
    
    finish = time.perf_counter()
 
    print(f'{round(finish-start,2)}초 만에 작업이 완료되었습니다.')
```
