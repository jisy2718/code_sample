```python
import glob
import pandas as pd
article_filename_lst = glob.glob('./data/article/*.csv')

article_df = pd.DataFrame()
for path_filename in article_filename_lst:
    cur_df = pd.read_csv(path_filename)
    article_df = pd.concat([article_df, cur_df])

a = ( article_df['from_kwd_name'].value_counts() > 4)
over_30 = set(a[a].index) 
over_30_article_df = article_df[article_df['from_kwd_name'].isin(over_30)]
over_30_article_df.sort_values(['from_kwd_name','pub_date'], ascending=False)
```

## [1] group 별로 상위 4개씩 가져오기

```python
 tmp = over_30_article_df.sort_values(by='pub_date', ascending=False).groupby('from_kwd_name', sort=False).head(4)  # DF의 GROUPBY해도 순서는 유지됨
 tmp[tmp['from_kwd_name']=='AI 반도체']
```
