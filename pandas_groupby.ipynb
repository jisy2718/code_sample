## [1] 성질

### (1) for loop으로 접근
+ 해당 group의 df를 내보낸다

```python
import pandas as pd
test = pd.DataFrame(data={'user':[1,1,1,1,2,2,3,4,5,6,6], 'col1':list(range(len([1,1,1,1,2,2,3,4,5,6,6])))})

test_group = test.groupby('user')

for user, group in test_group:
    print(user)
    print(group)
    print()

```

## [2] 활용

### (1) user_id 별로 unique한 user_session이 몇 개 인지 세기


```python
import pandas as pd
a = pd.DataFrame({'user_id':[1,1,1,2,2,2,2,2,2,3,3,3,4] , 'user_session':['a','a','a','b','b','c','d','e','f','z','z','y','k']})
print(a)
```

```python
c = a.groupby('user_id')['user_session'].unique()
print(c)
```

```python
b = a.groupby('user_id')['user_session'].nunique()
print(b)
```

```python
b.value_counts()  # 1, 5, 2가 각각 1번씩 등장
```

```python
# .items() 로 index와 value 모두 접근 가능
for idx, value in b.items():
    print(idx, value)
```
